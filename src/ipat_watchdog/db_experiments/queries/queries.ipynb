```python
from kadi_apy import KadiManager

m = KadiManager()

# Simple search by identifier string
resp = m.search.search_resources(
    "record",
    q="identifier:psa_01-jfi-ipat-h_test1"
)

hits = resp.json()
print(hits)
```

```python
resp = m.make_request(
    "https://134.169.58.190/api/v1/records",
    method="get",
    params={"identifier": "psa_01-jfi-ipat-h_test1"}
)
data = resp.json()
print(data)
```

```python
from kadi_apy import KadiManager
import json

m = KadiManager()

# Use a wildcard search in the query string
resp = m.search.search_resources("record", q="identifier:psa_*")

records = resp.json()
print(json.dumps(records, indent=2))
```

```python
from kadi_apy import KadiManager

m = KadiManager()
resp = m.make_request(
    "https://134.169.58.190/api/v1/records",
    method="get",
    params={
        "query": "identifier:psa_*",   # prefix match
        "per_page": 100,
        "sort": "identifier"
    }
)
data = resp.json()

print(json.dumps(data, indent=2))
```
